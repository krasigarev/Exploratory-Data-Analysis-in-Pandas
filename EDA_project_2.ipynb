## Data Preperation

```python
df.head(1)
```

```python
df = df[[
    'coaster_name',
    'Location',
    'Status',
    'Manufacturer',
    'year_introduced',
    'latitude',
    'longitude',
    'Type_Main',
    'opening_date_clean',
    'speed_mph',
    'height_ft',
    'Inversions_clean',
    'Gforce_clean'
]]

df
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df = df.rename(columns={'coaster_name':'Coaster_Name',
                   'year_introduced':'Year_Introduced',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_mph',
                   'height_ft':'Height_ft',
                   'Inversions_clean':'Inversions',
                   'Gforce_clean':'Gforce'
})
```

```python
df.isna().sum()
```

```python
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location', 'Opening_Date'])].reset_index(drop=True).copy()
```

## Feature Understanding

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Years Coasters Introduced')

ax.set_xlabel('Year Introduced')
ax.set_ylabel('Count')
```

```python
ax = df['Speed_mph'].plot(kind='hist', bins=20, title='Coaster Speed (mph)')
ax.set_xlabel('Speed mph')
```

```python
ax = df['Speed_mph'].plot(kind='kde', title='Coaster Speed (mph)')
ax.set_xlabel('Speed (mph)')
```

```python
df['Type_Main'].value_counts()
```

## Feature Relationships

```python
df.plot(kind='scatter',
        x='Speed_mph',
        y='Height_ft',
        title='Coaster Speed vs. Height')
plt.show()
```

```python
ax = sns.scatterplot(x='Speed_mph',
                   y='Height_ft',
                   hue='Year_Introduced',
                   data=df)
ax.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(df, vars=['Year_Introduced', 'Speed_mph', 'Height_ft','Inversions', 'Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Height_ft', 'Inversions', 'Gforce']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr, annot=True)
```

## Ask a Question about the data

What are the locations with the fastest roller coasters (minimum of 10)?

```python
ax = df.query('Location != "Order"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12,5), title='Average Coast Speed vs Location')
ax.set_xlabel('Average Coast Speed')
plt.show()
```
